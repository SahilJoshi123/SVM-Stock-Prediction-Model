Build the dataset according to the feature set:

```python
FEATURES =  ['DE Ratio',
             'Trailing P/E',
             'Price/Sales',
              'Price/Book',
             'Profit Margin',
              'Return on Equity',
             'Revenue Per Share',
             'Market Cap',
             'Enterprise Value',
             'Forward P/E',
             'PEG Ratio',
             'Enterprise Value/Revenue',
             'Enterprise Value/EBITDA',
             'Revenue',
             'Gross Profit',
             'EBITDA',
             'Net Income Avl to Common ',
             'Diluted EPS',
             'Earnings Growth',
             'Revenue Growth',
             'Total Cash',
             'Total Cash Per Share',
             'Total Debt',
             'Current Ratio',
             'Book Value Per Share',
             'Cash Flow',
             'Beta',
             'Held by Insiders',
             'Held by Institutions',
             'Shares Short (as of',
             'Short Ratio',
             'Short % of Float',
             'Shares Short (prior)'
             ]

def Build_Dataset():
    df = pd.read_csv('Processed_Data.csv')
    
    # To randomize the DataSet
    df = df.reindex(np.random.permutation(df.index))
    
    y = df['Status'].replace('Down',0).replace('Up',1).to_numpy()
    
    # To preprocess and standardize the training datset
    scaler = preprocessing.StandardScaler()
    X = scaler.fit_transform(df[FEATURES])
    
    
    return X,y
    
```

```python
def Analysis():
    
    test_size = 1000
    X, y = Build_Dataset()
    print(len(X), len(y))
    
    clf = svm.SVC(kernel='linear', C=1.0)
    clf.fit(X[:-test_size],y[:-test_size])
  
    
    count = 0
    
    for x in range(1,test_size+1):
        if clf.predict([X[-x]]) == y[-x]:
            count += 1

    print('Accuracy: ', (count/test_size)*100)
            
    
    #For Graphing the Model:  
#     w = clf.coef_[0]
#     a = -w[0]/w[1]
#     b = clf.intercept_[0]
    
#     xx = np.linspace(min(X[:, 0]), max(X[:, 0]))
#     yy = a*xx - b/w[1]
    
#     h0 = plt.plot(xx,yy, 'k-', label = 'Non Weighted')
    
#     plt.scatter(X[:, 0], X[:, 1], c=y)
#     plt.legend()
#     plt.ylabel('Trailing P/E')
#     plt.xlabel('DE Ratio')
#     plt.show()
    
    
```
